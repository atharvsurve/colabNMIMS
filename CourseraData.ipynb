```python
filtered_df['Tags'] =  filtered_df['Tags'].apply(lambda x:" ".join(x))
```

```python
filtered_df['Tags'] = filtered_df['Tags'].apply(lambda x : x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
cv
```

```python
vectors = cv.fit_transform(filtered_df['Tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
filtered_df['Tags'] = filtered_df['Tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
filtered_df.columns = ['Title' , 'Tags', 'Course Rating', 'URL' ]
filtered_df.columns
```

```python
def recommend(Course):
    filtered_courses = filtered_df[filtered_df['Title'] == Course]

    if not filtered_courses.empty:
        Course_index = filtered_courses.index[0]
        distances = similarity[Course_index]
        Course_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

        for i in Course_list:
            print(filtered_df.iloc[i[0]].Title)
            print(filtered_df.iloc[i[0]].URL)
    else:
        print(f"No matching course found for '{Course}'.")
```

```python
recommend('Finance for Managers')
```
